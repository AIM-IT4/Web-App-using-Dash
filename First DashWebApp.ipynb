```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
```

```python
data = pd.read_csv("Yearwise data.csv")



app = dash.Dash(__name__)
```

```python
data
```

```python
app.layout = html.Div(
    children=[
        html.H1(children="Area Irrigated",),
        html.P(
            children=" Analysing Year wise area irrigated by various techniques",
        ),
        dcc.Graph(
            figure={
                "data": [
                    {
                        "x": data["Year"],
                        "y": data["Area by tubewells"],
                        "type": "lines",
                    },
                ],
                "layout": {"title": "Area by tubewells"},
            },
        ),
        dcc.Graph(
            figure={
                "data": [
                    {
                        "x": data["Year"],
                        "y": data["Total"],
                        "type": "lines",
                    },
                ],
                "layout": {"title": "net area sown year wise "},
            },
        ),
    ]
)
```

```python
app.run_server(port = 8090, dev_tools_ui=True, #debug=True,
              dev_tools_hot_reload =True, threaded=True)
```
